```python
import numpy as np
import pandas as pd
```

```python
data1 = pd.read_csv('/Users/sdbgpro/Documents/THP/data/S2/mardi/data_cleaning/data/db_start/Invoice_20201220.csv', sep='|')

data1
```

```python
#Afficher le nom des antennes dans le format "4 chiffres et 2 lettres"

data1['SITE_NAME'] = data1['SITE_NAME'].str[11:17]

data1
```

```python
# supprimer la colonne "INVOICE_NAME" qui n'apporte rien aux Ã©quipes.

data1 = data1.drop(columns = 'INVOICE_NAME')

data1
```

```python
# CrÃ©er une colonne qui calcule le nombre de jours entre le dÃ©but et la fin de la pÃ©riode de facturation.

data1['TOTAL_DAYS'] = pd.to_datetime(data1['END']) - pd.to_datetime(data1['BEGIN'])

data1
```

```python
# CrÃ©er une autre colonne qui nous donne la consommation journaliÃ¨re moyenne de la pÃ©riode.

data1.dtypes
```

```python
data1['KWH_AVG_CONSO'] = (data1['KWH_PERIOD'].replace(' ','', regex=True).astype(str).astype(int)) / data1['TOTAL_DAYS'].dt.days

data1
```

```python
# CrÃ©er une colonne qui donne, pour chaque antenne, le nombre de factures dont on dispose au total.

data1['NB_INVOICES'] = data1.groupby('SITE_NAME')['SITE_NAME'].transform('size')

data1
```
